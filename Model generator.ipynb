```python
 from transformers import GPT2Tokenizer, AutoModel,GPT2Model, GPT2Config
```

```python
configuration = GPT2Config()
```

```python
model = GPT2Model.from_pretrained('gpt2')
```

```python
tokenizer = GPT2Tokenizer.from_pretrained("gpt2")
```

```python
numList=tokenizer("Hello world , this is david speaking", return_tensors="pt")['input_ids']
```

```python
outputs = model(numList)
```
